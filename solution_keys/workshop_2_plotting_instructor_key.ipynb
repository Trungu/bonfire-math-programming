## Brief SymPy review

Before we begin plotting, we will briefly review a few SymPy tools that will be useful later in the workshop.

First, import SymPy. The alias `sp` is a standard abbreviation that makes SymPy commands shorter and easy to recognize. This is the most commonly accepted convention and one you should use

````python
import sympy as sp
````

### Defining and evaluating an expression

First, define the symbol $x$ and a symbolic expression. The `.subs()` method allows us to substitute a value for $x$.

````python
x = sp.symbols("x")
review_function = x ** 2 + 2 * x + 1

display(review_function)
display(review_function.subs(x, 3))
````

### Differentiating an expression

The command `sp.diff(expression, variable)` calculates a derivative. We can apply it more than once to find higher-order derivatives.

````python
first_derivative = sp.diff(review_function, x)
second_derivative = sp.diff(review_function, x, 2)

display(first_derivative)
display(second_derivative)
````

### Calculating a limit

The command `sp.limit(expression, variable, value)` calculates the value that an expression approaches. For example, consider

$$\lim_{x \to 0}\frac{\sin(x)}{x}.$$

````python
limit_expression = sp.sin(x) / x
sp.limit(limit_expression, x, 0)
````

## Plotting functions with SymPy

SymPy can graph symbolic expressions using Matplotlib as its plotting backend. This allows us to create a graph directly from an expression without first calculating a list of numerical points.

We will use SymPy's `plot()` function throughout this workshop.

### Defining a function

Before graphing a function, we represent it as a SymPy expression. The following code defines the symbol $x$ and the expression

$$f(x) = x^2-8.$$

````python
x = sp.symbols("x")
f_x = (x ** 2) - 8
````

### Creating a basic plot

SymPy's `sp.plot()` function creates a graph from a symbolic expression. Its basic form is

```python
sp.plot(expression, (variable, minimum, maximum))
```

The two main arguments are:

- `expression`: the symbolic expression to graph
- `(variable, minimum, maximum)`: the independent variable and the interval displayed on the horizontal axis—that is, the plotting domain

SymPy evaluates the expression across the given interval and uses those values to draw the graph.

For our specific example, we want to graph $f(x) = x^2-8$ over the interval $-10 \leq x \leq 10$. Therefore, we pass `f_x` as the expression and `(x, -10, 10)` as the plotting interval.

````python
sp.plot(f_x, (x, -10, 10))
````

### Making a graph easier to read

A graph should clearly communicate what it represents. We can pass additional keyword arguments to `sp.plot()` to add a descriptive title, label the axes, select a line color, and control the visible vertical range.

Complete the blank attributes below as a class.

````python
sp.plot(
    f_x,
    (x, -10, 10),
    title = "Graph of f(x) = x² - 8",
    xlabel = "x",
    ylabel = "f(x)",
    line_color = "blue",
    ylim = (-10, 20)  # Limit the vertical axis.
)
````

In the code above:

- `title` describes the graph.
- `xlabel` and `ylabel` label the horizontal and vertical axes.
- `line_color` changes the color of the curve.
- `ylim` sets the portion of the vertical axis that is displayed.

The viewing window is an important mathematical choice. A poorly chosen interval can hide important behavior or make a graph difficult to interpret.

### Graphing multiple functions

We can also use `sp.plot()` to display multiple functions on the same coordinate plane. Comparing functions visually can help us investigate questions such as where two functions have the same value.

Consider the functions

$$f(x) = x^2-8 \qquad\text{and}\qquad g(x) = x+4.$$

Before running the next cell, predict how many times their graphs will intersect.

````python
# 1. Define the second function.
g_x = x + 4

# 2. Create a plot containing both functions.
# show = False allows us to customize the plot before displaying it.
p = sp.plot(
    f_x,
    g_x,
    (x, -10, 10),
    title = "Graphs of f(x) and g(x)",
    xlabel = "x",
    ylabel = "y",
    ylim = (-10, 20),
    legend = True,
    show = False
)

# 3. Set the color and legend label of each curve.
p[0].line_color = "blue"
p[1].line_color = "green"
p[0].label = "f(x)"
p[1].label = "g(x)"

# 4. Display the completed plot.
p.show()
````

The graph allows us to estimate the intersection points. We can verify their $x$-coordinates algebraically by setting $f(x) = g(x)$ and using SymPy's `solve()` function.

````python
# 1. Set f(x) equal to g(x) by moving everything to one side.
equation = f_x - g_x

# 2. Solve the resulting equation for x.
solutions = sp.solve(equation, x)

print(f"The solutions to the equation f(x) = g(x) are: {solutions}")
````

To compare the symbolic solutions with the graph, we can mark each solution with a vertical dashed line. SymPy creates the original function plots, and Matplotlib's `axvline()` command adds the vertical reference lines.

````python
# 1. Create the two function plots without displaying them yet.
p = sp.plot(
    f_x,
    g_x,
    (x, -10, 10),
    title = "Graphs of f(x) and g(x)",
    xlabel = "x",
    ylabel = "y",
    ylim = (-10, 20),
    legend = True,
    show = False
)

# 2. Prepare the underlying Matplotlib graph for customization.
p.process_series()

# 3. Draw a vertical dashed line at each solution.
for solution in solutions:
    p.ax.axvline(solution, color = "red", linestyle = "--")

# 4. Display the completed plot.
p.fig.tight_layout()
p.plt.show()
````

### Plotting a function and its derivatives

Plotting a function together with its derivatives can help us visualize how they are related. Recall that:

- The first derivative, $f'(x)$, describes the slope of $f(x)$.
- The second derivative, $f''(x)$, describes how the slope is changing and helps us identify concavity.

Consider the function

$$f(x) = x^4-4x^2.$$

We will use SymPy to calculate its first and second derivatives, then plot all three expressions on the same coordinate plane.

````python
# 1. Define the function and calculate its derivatives.
f_x = (x ** 4) - 4 * (x ** 2)
first_derivative = sp.diff(f_x, x)
second_derivative = sp.diff(f_x, x, 2)

# 2. Create a plot containing all three expressions.
p = sp.plot(
    f_x,
    first_derivative,
    second_derivative,
    (x, -3, 3),
    title = "f(x) and its derivatives",
    xlabel = "x",
    ylabel = "y",
    ylim = (-20, 40),
    legend = True,
    show = False
)

# 3. Give each curve a color and a descriptive label.
p[0].line_color = "blue"
p[0].label = "f(x)"

p[1].line_color = "green"
p[1].label = "f'(x)"

p[2].line_color = "orange"
p[2].label = "f''(x)"

# 4. Display the completed plot.
p.show()
````

## Group problem: From secant lines to a tangent line

A derivative can be understood as the limit of the slopes of secant lines. For a function $f(x)$, the difference quotient at $x = a$ is

$$\frac{f(a+h)-f(a)}{h}.$$

This expression gives the slope of the secant line through the points $(a,f(a))$ and $(a+h,f(a+h))$. As $h$ approaches zero, the second point moves toward the first and the secant line approaches the tangent line.

In this activity, your group will build the visualization one step at a time using $f(x) = x^2$ at $a = 1$. Work together on each part before moving to the next one. Be prepared to explain your group's reasoning during the class discussion.

### Part 1: Construct one secant line

Begin by defining $f(x) = x^2$, the fixed value $a = 1$, and $h = 2$. You will use these values to construct one secant line.

````python
# Define the function and the fixed value a.
f_x = x ** 2
a = 1
h = 2

# Complete the two function values.
first_y = f_x.subs(x, a)
second_y = f_x.subs(x, a + h)
````

Use the values above to calculate the secant slope and construct the secant line. The slope and point-slope formulas are provided.

````python
# Calculate the slope of the secant line.
secant_slope = (second_y - first_y) / h

# Use point-slope form to construct the secant line.
secant_h2 = first_y + secant_slope*(x - a)

print(f"Secant slope: {secant_slope}")
print(f"Secant line: {secant_h2}")
````

### Part 2: Define and test a reusable function

Repeating the calculation for every value of $h$ would be inefficient. Create a Python function named `secant_line(h)` that performs the calculation for any nonzero value of $h$.

Inside the function, your group will need to write two expressions:

1. Calculate the secant slope using the difference quotient

$$m = \frac{f(a+h)-f(a)}{h}.$$

2. Construct the secant line using point-slope form through the fixed point $(a,f(a))$

$$y = f(a)+m(x-a).$$

The completed SymPy translation is shown below.

````python
def secant_line(h):
    # Calculate the slope using the difference quotient.
    slope = (f_x.subs(x, a + h) - f_x.subs(x, a)) / h

    # Use point-slope form to construct the line.
    line = f_x.subs(x, a) + slope * (x - a)
    return line
````

Test the function using several values of $h$. Examine the resulting equations and look for a pattern.

````python
secant_h2 = secant_line(2)
secant_h1 = secant_line(1)
secant_h05 = secant_line(sp.Rational(1, 2)) # keeps the fraction in exact form

print(f"When h = 2:   {secant_h2}")
print(f"When h = 1:   {secant_h1}")
print(f"When h = 1/2: {secant_h05}")
````

### Part 3: Plot and interpret the secant lines

Now plot $f(x)$ and the three secant lines on the same coordinate plane. Most of the formatting is provided so that your group can focus on the mathematics.

Before running the cell, predict how the lines will move as $h$ gets closer to zero.

````python
# Plot the function and the secant lines.
p = sp.plot(
    f_x,
    secant_h2,
    secant_h1,
    secant_h05,
    (x, -1, 4),
    title = "Secant Lines Approaching a Tangent Line",
    xlabel = "x",
    ylabel = "y",
    ylim = (-2, 12),
    legend = True,
    show = False
)

# Give each curve a color and a descriptive label.
p[0].line_color = "black"
p[0].label = "f(x) = x²"

p[1].line_color = "red"
p[1].label = "h = 2"

p[2].line_color = "orange"
p[2].label = "h = 1"

p[3].line_color = "green"
p[3].label = "h = 1/2"

# Display the completed plot.
p.show()
````

### Part 4: Compare with the tangent line

The derivative $f'(a)$ gives the exact slope of the tangent line at $x = a$. The completed calculation is shown below.

````python
# Differentiate f(x).
derivative = sp.diff(f_x, x)

# Evaluate the derivative at x = a.
tangent_slope = derivative.subs(x, a)

# Construct the tangent line using point-slope form.
tangent_line = f_x.subs(x, a) + tangent_slope * (x - a)

print(f"Derivative: {derivative}")
print(f"Tangent slope at x = {a}: {tangent_slope}")
print(f"Tangent line: {tangent_line}")
````

Run the final graph and compare the secant line for $h = \frac12$ with the exact tangent line.

````python
# Plot the function, closest secant line, and tangent line.
p = sp.plot(
    f_x,
    secant_h05,
    tangent_line,
    (x, -1, 4),
    title = "Comparing a Secant Line and the Tangent Line",
    xlabel = "x",
    ylabel = "y",
    ylim = (-2, 12),
    legend = True,
    show = False
)

p[0].line_color = "black"
p[0].label = "f(x) = x²"

p[1].line_color = "green"
p[1].label = "Secant line: h = 1/2"

p[2].line_color = "blue"
p[2].label = "Tangent line"

# Display the completed plot.
p.show()
````

## Appendix: SymPy plotting quick reference

Use this section as a short reference when creating your own plots.

### Plot one expression

```python
sp.plot(expression, (variable, minimum, maximum))
```

Example:

```python
sp.plot(x ** 2, (x, -5, 5))
```

### Plot multiple expressions

Pass each expression as a separate argument:

```python
sp.plot(f_x, g_x, (x, -5, 5), legend = True)
```

### Common plot options

```python
sp.plot(
    f_x,
    (x, -5, 5),
    title = "My Graph",
    xlabel = "x",
    ylabel = "f(x)",
    line_color = "blue",
    ylim = (-10, 10)
)
```

- `title`: adds a title
- `xlabel`, `ylabel`: label the axes
- `line_color`: changes the curve's color
- `xlim`, `ylim`: control the visible plotting window
- `legend = True`: displays a legend when plotting multiple expressions

### Customize multiple curves

Use `show = False` to create a plot before displaying it:

```python
p = sp.plot(f_x, g_x, (x, -5, 5), legend = True, show = False)

p[0].line_color = "blue"
p[0].label = "f(x)"
p[1].line_color = "green"
p[1].label = "g(x)"

p.show()
```

### Other useful tools

These four commands are especially useful for investigating and verifying calculus problems:

```python
# Simplify an expression
sp.simplify(expression)

# Solve an equation for x
sp.solve(equation, x)

# Calculate a limit as x approaches a
sp.limit(expression, x, a)

# Differentiate with respect to x
sp.diff(expression, x)
```

> Remember: a graph can help reveal and explain a pattern, while symbolic tools can verify results exactly.
