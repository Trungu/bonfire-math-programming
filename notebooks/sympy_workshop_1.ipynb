### Introduction to SymPy

SymPy is a Python library. A library is essentially a set of external tools that we may use to assist us in our program. We must first _import_ SymPy.

```python
# IMPORTANT: run this cell to import the sympy library
import sympy
```

Variables can be represented in SymPy using `sympy.symbols(variable_name)`, where `variable_name` is the name of your variable.

```python
x = sympy.symbols("x")
y = sympy.symbols("y")

# you may also specify "assumptions" about a variable using the positive/negative arguments
# setting positive=True allows SymPy to ASSUME that z is positive, which will affect how
# it computes or simplifies certain expressions. For example square roots
z = sympy.symbols("z", positive=True)
```

Now we can do math operations with these variables

```python
c = (2 * x * y * z) + x * y + x ** 2

c 
```

> Note: we can display a value or expression in a notebook cell by simply putting it on the last line of the cell, as we have done when we called `c`. However, we can only display one value or expression per cell. If you want to display multiple values, you will need to use print()

This is cool, but the real merit in using SymPy is being able to do operations we can't normally do in vanilla Python. We can represent `mathematical functions` in python using the same syntax we used above.

```python
f_x = x ** 2 + 2 * x + 1 # this is a function of x, which we can call f(x) = x^2 + 2x + 1
g_xy = x ** 2 + y ** 2 # this is a function of x and y, which we can call g(x, y) = x^2 + y^2

# f_x is a function, so we can evaluate it at a specific value of x. For example, let's evaluate f(3):

f_3 = f_x.subs(x, 3) # this substitutes x with 3 in the function f_x
print(f"The value of f(3) is: {f_3}")

# we can sub in multiple values too
# this is done through the use of a dictionary
g_3_4 = g_xy.subs({x: 3, y: 4}) # this substitutes x with 3 and y with 4 in the function g_xy
print(f"The value of g(3, 4) is: {g_3_4}")
```

You can also substitute variables and expressions into other functions

```python
# we can define a variable d = 2x + 1
d = 2 * x + 1
f_d = f_x.subs(x, d) # this substitutes x with d in the function f_x

f_d
```

SymPy allows you to utilize various mathematical functions, such as trigonometry. You also have the ability to utilize certain constants, and even infinity!

```python
# some trig functions you can use
a = 5 * sympy.sin(x)
b = 3 * sympy.cos(x)
c = 8 * sympy.tan(x)

# using pi to calculate area of a circle.
# note: .evalf() evaluates an expression into a floating point number
radius = 5
circle_area = sympy.pi * x ** 2
result = circle_area.subs(x, radius)
print(f'The area of a circle with radius {radius} is {result} ({result.evalf()})') # .evalf() is used to force pi*5**2 to be a float

# exponential function e^x
exponential_fn = sympy.exp(x)
display(exponential_fn)

# using the constant e by itself with no exponential
print(f'The value of the constant e doubled equals {sympy.E * 2} ({(sympy.E * 2).evalf()})')
```

You can simplify expressions using `sympy.simplify()`. We can simplify a complex rational like $\frac{x^3 + x^2 - x - 1}{x^2 + 2x + 1}$

```python
# f(x) = (x^3 + x^2 - x - 1) / (x^2 + 2x + 1)
f_x =(x**3 + x**2 - x - 1) / (x**2 + 2*x + 1)

sympy.simplify(f_x)
```

We can also simplify the difference quotient with `sympy.simplify()`. This is useful in calculus because the difference quotient is the expression we simplify before taking a derivative.

```python
# Difference quotient example for f(x) = x^2
from IPython.display import Math

h = sympy.symbols("h")

# Display the default difference quotient formula before substituting a specific function
display(Math(r"\frac{f(x + h) - f(x)}{h}"))

f_x = x ** 2

# Difference quotient: (f(x + h) - f(x)) / h
difference_quotient = (f_x.subs(x, x + h) - f_x) / h
simplified_difference_quotient = sympy.simplify(difference_quotient)

display(difference_quotient)
display(simplified_difference_quotient)

# Taking the limit as h approaches 0 gives us the derivative
sympy.limit(simplified_difference_quotient, h, 0)
```

Some extra functions you may find useful when working with polynomials

```python
# Binomial and Polynomial Expansion using sympy.expand()
f_x = (2*x + 1) ** 2 # this is a binomial expression
f_x_expanded = sympy.expand(f_x) # this expands the binomial expression

display(f_x_expanded)
```

```python
# Polynomial Long Division using sympy.div()

g_x_dividend = (x**3 + 5*x**2 - 24 * x - 21)
g_x_divisor = (x**2 - 3*x)

quotient = sympy.div(g_x_dividend, g_x_divisor) # this performs long division of the dividend by the divisor

# the result of quotient is a tuple containing the quotient and remainder, we can do something called "tuple unpacking"
# which is a technique to extract values of a tuple.
# here, we display the quotient and remainder of the long division
print("The quotient of g(x) is: ")
display(quotient[0] + quotient[1] / g_x_divisor)
```

### Testing Multiple Values
Often it is helpful to plug various input values into a function to see what they equal. While you can do this manually, a loop can cut down the amount of code necessary significantly.

```python
f = sympy.exp(2 * x) - 4 * sympy.sin(x) # f(x) = e^(2x) - 4sin(x)

# we can manually plug in values
print(f"f(1) = {f.subs(x, 1)}") # substitute 1 for x
print(f"f(2) = {f.subs(x, 2)}") # now with 2
print(f"f(3) = {f.subs(x, 3)}") # now with 3
print(f"f(4) = {f.subs(x, 4)}") # now with 4

# separating new line between both sets of code
print()

# or we can use a loop to iterate over multiple values
for i in range(1, 5): # start at 1, and go in steps of 1 until you reach 4 (the top bound is excluded with range)
	print(f"f({i}) = {f.subs(x, i)}") # use the control variable to substitute for x in each iteration

# now imagine you had to check hundreds of values. which method is easier?
# how could you change the loop code so that you test all x values from 1 to 100?

# for simplicity this code block did not use .evalf(), but if you wanted, you can modify it to use .evalf().
# note how the output displays when you do not use .evalf() versus when you do
```

### Solving for x: Python Edition
SymPy can help you solve for x. And it's really simple! All you do is use the command **sympy.solve(function, variable)**, where *function* represents the equation in which you are trying to solve for a variable, and *variable* is what you are trying to solve for, in this case x! .solve() essentially sets your function equal to 0 to solve for the variable you are looking for.

```python
f = x ** 2 - 5
display(sympy.solve(f, x)) # note the output
```

### Special Case of Solving
Suppose you have some function, and you want to equate it to a specific value. For example, let's say you have a function which represents the volume of a box, and you want to figure out what side length of box you would need so that the volume equals 50. If you were to solve this by hand, you would probably take your volume function and set it equal to 50, and then you would solve for x. But .solve() sets your equation equal to 0. So how do we solve this in SymPy? Simple, just subtract the quantity on the other side of the equal sign to the other side. So in our box example, if we rearrange it to $volume - 50 = 0$, this is now an equation which SymPy can solve.

```python
desired_volume = 50

f = x ** 3 # volume of a cube
print(f"To achieve a volume of {desired_volume}, some possible side lengths include:")
print(sympy.solve(f - desired_volume, x))

# notice the result. this case has multiple answers, so you get a list of answers when running the code.
# some of them include an imaginary portion, or they are completely negative. from here you can pick the answer
# that makes the most sense. 
```
