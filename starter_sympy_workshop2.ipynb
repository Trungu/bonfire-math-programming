### Quick Review

Here is a quick review of the SymPy basics we need for today.

```python
x = sympy.symbols("x")
y = sympy.symbols("y")

f_x = x**2 + 2*x + 1
display(f_x)

print(f"f(3) = {f_x.subs(x, 3)}")
```

### Testing Multiple Values
Often it is helpful to plug various input values into a function to see what they equal. A loop can help avoid repeated code.

```python
# define a function f(x) = e^(2x) - 4sin(x)
f = sympy.exp(2 * x) - 4 * sympy.sin(x)

# plug in x = 1, 2, 3, and 4 manually
print(f"f(1) = {f.subs(x, 1)}")
print(f"f(2) = {f.subs(x, 2)}")
print(f"f(3) = {f.subs(x, 3)}")
print(f"f(4) = {f.subs(x, 4)}")

print() # blank line separation between previous output and next output

# plug in x = 1, 2, 3, and 4 with a loop
for i in range(1, 5):
    print(f"f({i}) = {f.subs(x, i)}")

print() # blank line separation

# same thing as above, but now evaluated as floats
for i in range(1, 5):
    print(f"f({i}) = {f.subs(x, i).evalf()}")
```

### Solving for x: Python Edition
SymPy can help solve equations using `sympy.solve(function, variable)`.

```python
f = x**2 - 5
solutions = sympy.solve(f,x)
display(solutions) # you may notice this produces 2 solutions when you run the code

# when we have multiple solutions, if we want to make them into floats, we have
# to apply .evalf() to each individual solution. you can do this manually using
# list indexing:
print("Solution #1:")
display(solutions[0], solutions[0].evalf()) # print original exact value, then float value

print("Solution #2:")
display(solutions[1], solutions[1].evalf()) # same thing as previous display()

# python starts counting at 0, which is why the first solution is index 0

print() # blank line separation

# the same thing can be done with a loop instead. this is VERY helpful for when you have a LOT of solutions
print("With loops now:")
for sol in solutions: # iterate through each solution
	display(sol.evalf())
```

### Special Case of Solving
If you want to solve for a variable when the function equals a specific value, subtract that value first so the equation equals 0.

```python
desired_value = 50
f = x**3

# this would solve for x in the equation x^3 - 50 = 0, which is equivalent to x^3 = 50.
# .solve() by default equates what you give it to 0.
print(sympy.solve(f - desired_value, x))
```

### Limits
We can compute limits in SymPy using the `sympy.limit()` function. You use the function like this: `sympy.limit(function, variable, c)` which effectively evaluates $\lim_{variable \to c} function$. Make sure to get the order of the parameters correct.

```python
f = sympy.sin(x) / x

display(sympy.limit(f, x, 0))
```

### Left & Right Hand Limits
We can also look at limits from the left or right.

```python
f = 1 / x

print(sympy.limit(f, x, 0, dir="-"))
print(sympy.limit(f, x, 0, dir="+"))
```

### Limits at Infinity
To compute a limit at infinity, use `sympy.oo` to represent infinity. Think of the 2 o's making an infinity symbol to remember this.

```python
f = x / sympy.sqrt(x**2 + 1)

print("Limit at positive infinity:")
display(sympy.limit(f, x, sympy.oo))

print("Limit at negative infinity:")
display(sympy.limit(f, x, -sympy.oo))
```

### Derivatives
SymPy can compute the derivative of a function. You can do this by using the `sympy.diff()` function. The function is used like this: `sympy.diff(function, variable)`. This will compute the derivative of `function` with respect to `variable`. For our purposes, we will only focus on single-variable functions. You will deal with multivariable in MATH 251 (Calculus 3).

```python
f = x**2 + 3*x + 1

# differentiate with respect to x
derivative = sympy.diff(f, x)

print("Derivative:")
display(derivative)
print() # blank line separator

# we can still use .subs() on our derivative
print("f'(5):")
display(derivative.subs(x, 5)) # this computes f'(5)
print() # blank line separator

# and we can still differentiate our derivative
print("f''(x):")
display(sympy.diff(derivative, x))

# the derivative is stored as a new function, so you can use sympy commands on it as you wish
```
